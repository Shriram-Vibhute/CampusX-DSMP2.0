```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
with open(file = "smartphones.html", mode = 'r', encoding = 'utf-8') as f:
    html = f.read()
    soup = BeautifulSoup(html, 'lxml')
```

```python
containers = soup.findAll('div', {'class': "sm-product has-tag has-features has-actions"})
len(containers)
```

```python
name = []
price = []
sim = []
processor = []
RAM = []
battery = []
display = []
camera = []
card = []
os = []

for phone in containers:
    specifications = phone.find('ul', {'class': "sm-feat specs"}).findAll('li')
    if len(specifications) == 8:
        sim.append(specifications[0].get_text())
        processor.append(specifications[1].get_text())
        RAM.append(specifications[2].get_text())
        battery.append(specifications[3].get_text())
        display.append(specifications[4].get_text())
        camera.append(specifications[5].get_text())
        card.append(specifications[6].get_text())
        os.append(specifications[7].get_text())

        name.append(phone.find('h2').get_text())
        price.append(phone.find('span', {'class': 'price'}).get_text())
```

```python
df = pd.DataFrame(
    {
        'Model': name,
        'Price': price,
        'Sim': sim,
        'Processor': processor,
        'RAM': RAM,
        'Battery': battery,
        'Display': display,
        'Camera': camera,
        'Card': card,
        'OS': os
    }
)
df.head()
```

```python
with open("smartphones.csv", "w", encoding = "utf-8") as f:
    df.to_csv(f, index = False)
```
