```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

### `Q-1` - Using `Gapminder` Data
Create a scatter plot showing the relationship between 'gdpPercap' and 'lifeExp' for the year 2007. Adjust the marker size based on 'pop' and color based on 'continent'.

```python
import plotly.express as px
df = px.data.gapminder()
df.head()
```

```python
plt.figure(figsize=(15, 7))
scatterplot = sns.scatterplot(
    data=df,
    x='gdpPercap',
    y='lifeExp',
    hue='continent',
    size='pop',
)
scatterplot.legend(loc='upper right', bbox_to_anchor=(1.15, 1), title='Continent')
plt.xlabel('GDP per Capital')
plt.ylabel('Life Expecations')
plt.title('Scatter Plot')
plt.show()
```

### `Q-2-3` - Utilizing the `flights` dataset provided by seaborn.

2. Employing the "flights" dataset inherent to seaborn, generate a heatmap illustrating the mean quantity of passengers per month for each year.

3. Leveraging seaborn's flight dataset, construct a clustermap to depict the correlation among the quantity of passengers, months, and year.

```python
df = sns.load_dataset('flights')
df.head()
```

```python
# Creating pivot table
plt.figure(figsize=(10, 10))
sns.heatmap(df.pivot_table(index = 'year', columns = 'month', values = 'passengers', aggfunc = 'mean', observed=True))
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.clustermap(df.pivot_table(index = 'year', columns = 'month', values = 'passengers', aggfunc = 'mean', observed=True))
plt.show()
```

### `Q-4` - Create a scatter plot with the following conditions:
1. Set "age" as the x-axis and "bmi" as the y-axis.
2. Use "diabetic" for hue, "gender" for size, and "smoker" for style.
3. Include a title for the chart.
4. Filter the data for ages below the 70th percentile.
5. Consider only BMI values above the average of the filtered age group.

```python
df = pd.read_csv("insurance_data - insurance_data.csv")
df.head()

# Filtering the dataset as per the required conditions
new_df = df[df['age'] >= np.quantile(df['age'].dropna(), 0.7)]
new_df = new_df[new_df['bmi'] >= new_df['bmi'].mean(numeric_only = True)]
new_df.head()
```

```python
plt.figure(figsize = (15, 7))
sns.scatterplot(
    data = new_df,
    x = 'age',
    y = 'bmi',
    hue = 'diabetic',
    size = 'gender',
    style = 'smoker'
)
plt.show()
```

### `Q-5` - Draw a line plot by using the below informations

1. bloodpressure vs children
2. Blood-pressure values should be between 90 and 100. The upper and lower limit are included.
3. Show the details of "smoker".

```python
# Filtering the Dataset
new_df = df[(df['bloodpressure'] >= 90) | (df['bloodpressure'] <= 100)]

sns.relplot(
    data=new_df,
    x='bloodpressure',
    y='children',
    kind='line',
    hue = 'smoker',
    height = 7,
    aspect = 2.5,
)
plt.show()
```

### `Q-6` - Draw a histogram using displot

- based on "age" column.
- Show details of "smoker" (hue).
- Create 2 separate charts for the above 2 conditions based on "gender" side-by-side.

```python
sns.displot(
    data = df,
    x = 'age',
    kind = 'hist',
    col = 'gender',
    hue = 'smoker',
    height = 7,
    aspect = 1.5
)
plt.show()
```

### `Q-8` - Draw a clustermap between between "age", "bmi" and "bloodpressure".

```python
df['bpcuts'] = pd.cut(df['bloodpressure'], bins = 6, labels = ['85', '95', '105', '115', '125', '135'])
sns.clustermap(
    df.dropna(axis = 0, how = 'any').pivot_table(index = 'age', columns = 'bpcuts', values = 'bmi', aggfunc = 'mean', observed = True).fillna(0)
)
plt.show()
```
