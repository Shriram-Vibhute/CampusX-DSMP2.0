```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

df = pd.concat([train_df.drop(columns=['Survived']),test_df]).sample(1309)

plt.figure(figsize=(9, 6))
sns.kdeplot(df['Fare'])
plt.show()
```

```python
# sample size = 50 -> 100 times
samples = []
for i in range(100):
    samples.append(df['Fare'].dropna().sample(150).values.tolist())
samples = np.array(samples)
sampling_means = samples.mean(axis=1)

plt.figure(figsize=(9, 6))
sns.kdeplot(sampling_means)
plt.show()
```

```python
print(sampling_means.mean())
print(sampling_means.std() / np.sqrt(150))
```

```python
# Confidience Interval - 95%
lower_limit = sampling_means.mean() - 1.96*sampling_means.std()
upper_limit = sampling_means.mean() + 1.96*sampling_means.std()

print('The range is',lower_limit,'-',upper_limit)
```
